#### Q1. NumPy kya hai?

NumPy Python ki numerical computing library hai jo arrays aur mathematical operations ke liye use hoti hai.

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5,6,7])
print(arr)
```

#### 2. Array Ki Type Dekhna

```python
arr = np.array([1,2,3,4,5,6,7])
print(type(arr))
```

#### 3. 2D Array

```python
arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(arr)
```

#### 4. Array Properties

```python
arr = np.array([[1,2,3],[4,5,6]])

print(arr.shape)
print(arr.size)
print(arr.ndim)
```

#### 5. Indexing

```python
arr = np.array([10,20,30,40])

print(arr[0])
print(arr[2])
```

#### 6. Slicing

```python
arr = np.array([10,20,30,40,50])

print(arr[1:4])
```
