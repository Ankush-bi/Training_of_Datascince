2. Differentiate between: 
● Series and DataFrame  
● loc[] and iloc[] 
functions in Pandas 

| Feature    | Series                        | DataFrame                                       |
| ---------- | ----------------------------- | ----------------------------------------------- |
| Definition | One-dimensional labeled array | Two-dimensional labeled table                   |
| Dimensions | 1D                            | 2D                                              |
| Structure  | Single column of data         | Multiple rows and columns                       |
| Data Types | Usually one data type         | Different columns can have different data types |
| Index      | Has an index                  | Has both row index and column names             |
| Example    | Marks of students             | Complete student record                         |


| Feature         | loc[]                             | iloc[]                              |
| --------------- | --------------------------------- | ----------------------------------- |
| Selection Basis | Label (name) based                | Integer position based              |
| Rows/Columns    | Uses row and column labels        | Uses row and column index positions |
| End Index       | Included                          | Excluded (Python slicing rule)      |
| Syntax          | `df.loc[row_label, column_label]` | `df.iloc[row_index, column_index]`  |

  4. Create a DataFrame using the dataset below:
   Name A B C D 
   a) Tasks: Age 25 NaN 27 24 
   ● Identify missing values  Salary 30000 40000 NaN 35000 
   ● Fill missing Age with average age  
   ● Fill missing Salary with median salary  
   ● Display cleaned dataset  
   b) Expected Concepts: 
   ● isnull()  ● fillna()  ● mean()  ● median() 

| Function   | Purpose                 |
| ---------- | ----------------------- |
| `isnull()` | Detect missing values   |
| `fillna()` | Fill missing values     |
| `mean()`   | Calculate average value |
| `median()` | Calculate middle value  |


```python
data = {
      "Name": ["A", "B", "C", "D"],
    "Age": [25, np.nan , 27, 24],
    "Salary": [30000, 40000, np.nan , 35000]
}

df = pd.DataFrame(data)
print("Original Dataset:")
print(df)
```

```python
#Identify Missing Values
df.isnull()
```

```python
df.isnull().sum()
```

```python
avg_age = df["Age"].mean()
print(avg_age)
```

```python
# Fill Missing Age
df["Age"] = df["Age"].fillna(avg_age)
```

```python
median_salary = df["Salary"].median()
print(median_salary)
```

```python
df["Salary"] = df["Salary"].fillna(median_salary)
```

```python
print(df)
```
