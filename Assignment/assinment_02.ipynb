3. Check whether a number is an Armstrong number

```python
num = int(input("Enter a number: "))

temp = num
sum = 0

while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp = temp // 10

if sum == num:
    print(num, "is an Armstrong Number")
else:
    print(num, "is not an Armstrong Number")

```

6. Count the frequency of each word in a sentence using a dictionary.

```python
sentence = input("Enter the sentemce is ")
words = sentence.split()
frequency = {}
for word in words:
    if word in frequency:
        frequency[word]+=1
    else:
        frequency[word]=1 


print("Word Frequencies:")
for word, count in frequency.items():
    print(word, ":", count)
```

7. Display students scoring above the average marks from a dictionary.

```python
student = {
    "prishant": 50,
    "kalpesh": 80,
    "lakshit": 70,
    "ankush": 60,
    "bittu": 90,
    "ganesh": 60
}
sum = 0
count = 0
for name, marks in student.items():
    sum += marks
    count += 1
average = sum / count
print(student)
print(f"The average mark is: {average}")
print("Students scoring above average:")
for name, marks in student.items():
    if marks > average:
        print(f"Student: {name} with Marks: {marks}")

```

8. Create a menu-driven product management system using a dictionary.

```python
products = {}

while True:
    print("===== Product Management System =====")
    print("1. add product")
    print("2. view products")
    print("3. update product")
    print("4. delete product")
    print("5. search product")
    print("6. exit")

    choice = int(input("enter your choice: "))

    if choice == 1:
        name = input("enter product name: ")
        price = float(input("enter product price: "))
        products[name] = price
        print("product added successfully!")

    elif choice == 2:
        if products:
            print("\nproduct List:")
            for name, price in products.items():
                print(f"{name} : ₹{price}")
        else:
            print("no products available.")

    elif choice == 3:
        name = input("enter product name to update: ")
        if name in products:
            new_price = float(input("enter new price: "))
            products[name] = new_price
            print("product updated successfully!")
        else:
            print("product not found.")

    elif choice == 4:
        name = input("enter product name to delete: ")
        if name in products:
            del products[name]
            print("product deleted successfully!")
        else:
            print("product not found.")

    elif choice == 5:
        name = input("enter product name to search: ")
        if name in products:
            print(f"{name} : ₹{products[name]}")
        else:
            print("product not found.")

    elif choice == 6:
        print("Exiting Product Management System...")
        break

    else:
        print("invalid choice.")
```

9. Check whether a triangle is Equilateral, Isosceles, or Scalene. 

```python
a = int(input("Enter first side: "))
b = int(input("Enter second side: "))
c = int(input("Enter third side: "))

if a == b and b == c:
    print("Equilateral Triangle")

elif a == b or b == c or a == c:
    print("Isosceles Triangle")

else:
    print("Scalene Triangle")
```
