```python
point = np.array([1,2,3,4])
labal=np.array(["co2","h2o","o2","others"])
color = np.array(["red","black","blue"])
plt.pie(point,labels=labal,colors=colour,shadow=True)
plt.show()
```

```python
point = np.array([1,2,3,4])
labal=np.array(["co2","h2o","o2","others"])
color = np.array(["red","black","blue"])
x = np.array([0,0.2,0,0])
plt.pie(point,labels=labal,colors=colour,shadow=True,explode=x)
plt.legend()
plt.show()
```
