#  What is Exception Handling in Python?

Exception Handling is a way to handle errors that occur while a program is running, so the program doesn't crash unexpectedly.

```python
try: 
    \# Main code block 
except: 
    \# Code to handle exceptions
else: 
    \# Executes if no exception occurs 
finally: 
    \# Always runs, no matter what! 
try : Code that might throw an error. 
except : Runs if an error occurs. 
else : Runs if no error occurs. 
finally : Runs no matter what (even if you return or error out).
```

| Error Type              | Example Code                    | Description                                                                      |
| ----------------------- | ------------------------------- | -------------------------------------------------------------------------------- |
| **ZeroDivisionError**   | `10 / 0`                        | Occurs when dividing a number by zero.                                           |
| **ValueError**          | `int("abc")`                    | Occurs when a function receives a value of the correct type but invalid content. |
| **TypeError**           | `"10" + 5`                      | Occurs when an operation is performed on incompatible data types.                |
| **IndexError**          | `lst=[1,2,3]; lst[5]`           | Occurs when accessing a list index that does not exist.                          |
| **KeyError**            | `d={"name":"Ankush"}; d["age"]` | Occurs when accessing a dictionary key that does not exist.                      |
| **NameError**           | `print(age)`                    | Occurs when a variable is used before being defined.                             |
| **FileNotFoundError**   | `open("data.txt")`              | Occurs when trying to open a file that does not exist.                           |
| **AttributeError**      | `"hello".append("a")`           | Occurs when an object does not have the requested attribute or method.           |
| **ImportError**         | `import abcxyz`                 | Occurs when Python cannot import a module.                                       |
| **ModuleNotFoundError** | `import mymodule`               | Occurs when the specified module is not found.                                   |
