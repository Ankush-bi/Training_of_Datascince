write a function max_of_three(a,b,c)

```python
def max_of_three(a,b,c):
    if a >= b and a>= c:
        return a
    elif b>=c:
        return b
    else:
        return c
max_of_three(1,2,34)
```

```python
def calculator(a,b,operator):
    # operator = input("Enter the number: -")
    if operator =="+":
        return a+b
    elif operator =="-":
        return a-b
    elif operator =="*":
        return a*b
    elif operator =="/":
        return a/b
    else:
        return "invalid"
print(calculator(2,4,'+'))
print(calculator(2,4,'-'))
print(calculator(2,4,'*'))
print(calculator(2,4,'%'))

```

```python
# def factorial(n):
#     if n == 0 and n ==1:
#         return 1
#     else:
#         return n*factorial(n-1)
# print(factorial(5))
    
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))
    
```
