1. single Inhritance

```python
class company:
    company_ip = "34"
    company_name = "xyz"

class Developer(company):
    Developer_name = "name"
    Developer_no = "234578"

bittu = Developer()
print(bittu.company_ip)
print(bittu.Developer_name)
```

2. multiple Inhritance

```python
#multiple
class company:
    company_ip = "34"
    company_name = "xyz"

class Developer(company):
    Developer_name = "name"
    Developer_no = "234578"

class intern(Developer):
    intern_name = "ankush"
    inter_no = "3456"

bittu = intern()
print("this is main comany class")
print(bittu.company_ip)

print("this is main developer class")
print(bittu.Developer_name)

print("this is main intern class")
print(bittu.intern_name)
```

```python
class Father:
    def father_property(self):
        print("Father Property")

class Mother:
    def mother_property(self):
        print("Mother Property")

class Child(Father, Mother):
    pass

obj = Child()
obj.father_property()
obj.mother_property()
```

```python
#3. Multilevel Inheritance
class Grandfather:
    def house(self):
        print("Grandfather House")

class Father(Grandfather):
    pass

class Son(Father):
    pass

obj = Son()
obj.house()
```

```python
#4. Hierarchical Inheritance
class Parent:
    def show(self):
        print("Parent Class")

class Child1(Parent):
    pass

class Child2(Parent):
    pass

obj1 = Child1()
obj2 = Child2()

obj1.show()
obj2.show()
```

```python
#5. Hybrid Inheritance
class A:
    def showA(self):
        print("Class A")

class B(A):
    def showB(self):
        print("Class B")

class C(A):
    def showC(self):
        print("Class C")

class D(B, C):
    def showD(self):
        print("Class D")

obj = D()

obj.showA()
obj.showB()
obj.showC()
obj.showD()
```

### Private Attributes and Methods in Python

#### Python me private attributes aur private methods kisi class ke data ko direct access se bachane ke liye use kiye jate hain.

1. Private Attribute

   Variable ke naam ke aage __ (double underscore) lagate hain.

```python
class Student:
    def __show(self):      # Private method
        print("Private Method")

    def display(self):
        self.__show()      # Class ke andar access kar sakte hain

s = Student()
s.display()

# s.__show()  # Error
```

```python
class Car:
    color = "black"
    @staticmethod
    def start():
        print("\ncar started")

    @staticmethod
    def stop():
        print("car stop..")

class Toyota(Car):
    def __init__(self,name):
        self.name = name

car1 = Toyota("BMW")
car2 = Toyota("spot")

# print(car1.start())
print(car1.color)
print(car2.name)
```
