1. single Inhritance

```python
class company:
    company_ip = "34"
    company_name = "xyz"

class Developer(company):
    Developer_name = "name"
    Developer_no = "234578"

bittu = Developer()
print(bittu.company_ip)
print(bittu.Developer_name)
```

2. multiple Inhritance

```python
#multiple
class company:
    company_ip = "34"
    company_name = "xyz"

class Developer(company):
    Developer_name = "name"
    Developer_no = "234578"

class intern(Developer):
    intern_name = "ankush"
    inter_no = "3456"

bittu = intern()
print("this is main comany class")
print(bittu.company_ip)

print("this is main developer class")
print(bittu.Developer_name)

print("this is main intern class")
print(bittu.intern_name)
```
