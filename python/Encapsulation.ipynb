# Encapsulation


```python
class college:
    college_name = "anand"   #puclic
    _college_libary = "anand libary"#protected
    __college_account = 3456  #private
    def show(self):
        print(self.__college_account)


student = college()
print(student.college_name)
print(student._college_libary)
print(student.__college_account)
```

```python
# __a  = 10
class A:
    _a = 10
class B(A):
    pass

obj = B()
obj._a
```
