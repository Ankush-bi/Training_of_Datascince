### _ _init_ _ Function

#### All classes have a function called __init__(), which is always executed when the object is beinginitiated.

```python
class student:

    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
        print("New student name and marks is :--")

s1 = student("ankush",23)
print(s1.name,s1.marks)
s2 = student("bittu",30)
print(s2.name,s2.marks)
```
