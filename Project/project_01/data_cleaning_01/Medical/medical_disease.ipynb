```python
int_df = df.copy()
```

```python
int_df['Gender'] = int_df['Gender'].map({
    'Male':0,
    'Female':1
})

int_df['Smoking'] = int_df['Smoking'].map({
    'No':0,
    'Yes':1
})

int_df['Alcohol'] = int_df['Alcohol'].map({
    'No':0,
    'Yes':1
})

int_df['FamilyHistory'] = int_df['FamilyHistory'].map({
    'No':0,
    'Yes':1
})
```

```python
corr_matrix = int_df.corr(numeric_only=True)

print(corr_matrix)
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```
