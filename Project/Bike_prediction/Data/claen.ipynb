```python
brand_dict = {
    'TVS':1,   'Royal Enfield':2,         'Triumph':3,          'Yamaha':4,
           'Honda':5,            'Hero':6,           'Bajaj':7,          'Suzuki':8,
         'Benelli':9,             'KTM':10,        'Mahindra':11,        'Kawasaki':12,
          'Ducati':13,         'Hyosung':14, 'Harley-Davidson':15,            'Jawa':16,
             'BMW':17,          'Indian':18,         'Rajdoot':19,             'LML':20,
           'Yezdi':21,              'MV':22,           'Ideal':23
              }

df['brand']=df['brand'].map(brand_dict)
```

# # SVM (SVR)

```python
from sklearn.svm import SVC,SVR
sv=SVR()
sv.fit(x_train,y_train)
```

```python
pred=sv.predict(x_test)
pred
```

```python

print('SVR test score:-',sv.score(x_test,y_test))
print('SVR train score:-',sv.score(x_train,y_train))
```

# # SVM (SVC)

```python
from sklearn.svm import SVC
sv = SVC()
sv.fit(x_train,y_train)
```

```python
pred=sv.predict(x_test)
print(pred)


print('SVC test score:-',sv.score(x_test,y_test))
print('SVC train score:-',sv.score(x_train,y_train))
```

# # Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
pred
```

```python
print('test score:-',lr.score(x_test,y_test))
print('train score:-',lr.score(x_train,y_train))
```

# logisticRegression


```python
from sklearn.linear_model import LogisticRegression
log =LogisticRegression()
log.fit(x_train,y_train)
```

```python
print("LogisticRegression training",log.score(x_train,y_train))
print("LogisticRegression testing",log.score(x_test,y_test))
```

# # DecisionTreeClassifier


```python
from sklearn.tree import DecisionTreeClassifier
dt =DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python

print('test score:-',dt.score(x_test,y_test))
print('train score:-',dt.score(x_train,y_train))
```

# with perameter


```python
dt.get_params()
```

```python
from sklearn.tree import DecisionTreeClassifier
dt =DecisionTreeClassifier(criterion='entropy', min_samples_split=3, min_samples_leaf=2, random_state=42)
dt.fit(x_train, y_train)

```

```python
pred=dt.predict(x_test)
print(pred)
print('DecisionTreeClassifier test score:-',dt.score(x_test,y_test))
print('DecisionTreeClassifier train score:-',dt.score(x_train,y_train))
```

#  DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
import joblib
dt =DecisionTreeRegressor()
dt.fit(x_train, y_train)
```

```python
print('test score:-',dt.score(x_test,y_test))
print('train score:-',dt.score(x_train,y_train))
```

```python
joblib.dump(dt, 'DecisionTreeRegressor_model.lb')
print("Model saved successfully!")
```

```python
model=joblib.load('DecisionTreeRegressor_model.lb')
print(model.predict(x_test))
```
