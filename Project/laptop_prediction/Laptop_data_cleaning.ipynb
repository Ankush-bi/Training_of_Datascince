```python
df.drop_duplicates(inplace=True)
```

```python
df.drop('laptop_ID',axis=1,inplace=True)
df.drop
```

```python
df=df.drop_duplicates()
```

```python
for col in df.columns:
    print("\ncolumn:",col)
    print(df[col].unique())
```

```python
Company_list = df['Company'].unique().tolist()
Company_dict = {Company: i+1 for i, Company in enumerate(Company_list)}
print(Company_dict)
df['Company'] = df['Company'].map(Company_dict)
```

```python
Product_list = df['Product'].unique().tolist()
Product_dict = {Product: i+1 for i, Product in enumerate(Product_list)}
print(Product_dict)
df['Product'] = df['Product'].map(Product_dict)
```

```python
ScreenResolution_list = df['ScreenResolution'].unique().tolist()
ScreenResolution_dict = {ScreenResolution: i+1 for i, ScreenResolution in enumerate(ScreenResolution_list)}
print(ScreenResolution_dict)
df['ScreenResolution'] = df['ScreenResolution'].map(ScreenResolution_dict)
```

```python
TypeName_dict={
    'Ultrabook':1,           'Notebook':2,            'Netbook':3,
             'Gaming':4, '2 in 1 Convertible':5,        'Workstation':6
}
df['TypeName'] = df['TypeName'].map(TypeName_dict)

TypeName_dict
```

```python
df['Ram'] = df['Ram'].str.replace('GB', '').astype(int)
```

```python
Memory_list = df['Memory'].unique().tolist()
Memory_dict = {Memory: i+1 for i, Memory in enumerate(Memory_list)}
print(Memory_dict)
df['Memory'] = df['Memory'].map(Memory_dict)
```

```python
Gpu_list = df['Gpu'].unique().tolist()
Gpu_dict = {Gpu: i+1 for i, Gpu in enumerate(Gpu_list)}
print(Gpu_dict)

df['Gpu'] = df['Gpu'].map(Gpu_dict)
```

```python
Cpu_list = df['Cpu'].unique().tolist()
Cpu_dict = {Cpu: i+1 for i, Cpu in enumerate(Cpu_list)}
print(Cpu_dict)

df['Cpu'] = df['Cpu'].map(Cpu_dict)
```

```python
OpSys_list = df['OpSys'].unique().tolist()
OpSys_dict = {OpSys: i+1 for i, OpSys in enumerate(OpSys_list)}
print(OpSys_dict)
df['OpSys'] = df['OpSys'].map(OpSys_dict)
```

```python
df['Weight'] = df['Weight'].str.replace('kg', '').astype(float)
```

# SVM(SVR)

# DecisionTreeClassifier

```python

from sklearn.tree import DecisionTreeClassifier
dt =DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
pred=dt.predict(x_test)
print(pred)
print('DecisionTreeClassifier test score:-',dt.score(x_test,y_test))
print('DecisionTreeClassifier train score:-',dt.score(x_train,y_train))
```
